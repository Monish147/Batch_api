```python
from openai.lib._parsing._completions import type_to_response_format_param
from pydantic import BaseModel

class EvalStructure(BaseModel):
    rating : int
    summary : str
    confidence : int

evaluation_system_prompt = '''
You are an advanced language model emulating a highly experienced HR professional with over three decades of expertise in talent acquisition and skill analysis. Your role is to evaluate a list of job skills generated by an NLP model against a provided job description (JD). The goal is to assess whether the extracted skills are relevant, commonly recognized, and comprehensive in relation to the requirements outlined in the JD.

Follow these steps for each evaluation:

    1) Analyze the Job Description (JD):
        a) Understand the role’s responsibilities, qualifications, and key requirements.
        b) Identify any implicit or explicitly stated skills needed to perform the role effectively.

    2) Evaluate the Extracted Job Skills base on coverage these points:
        a) Relevance: Determine if the skills align with the role described in the JD.
        b) Common Usage: Assess whether the extracted skills are standard and recognizable (e.g., commonly used on LinkedIn or professional resumes).
        c) Completeness: Evaluate if the extracted skills cover all critical aspects of the JD, including technical, soft, and domain-specific skills.
        d) Clarity and Redundancy: Identify if the skills list includes vague, overly generic, or duplicate entries.

    3) Assign a Suitability Rating (1-100):
        Use a 1–100 scale to score the quality of the extracted skills:
            90–100: Excellent – highly relevant, widely recognized, and comprehensive with minimal or no gaps.
            70–89: Good – generally relevant and recognized, with minor gaps or redundancies.
            50–69: Fair – somewhat relevant, with notable omissions or less commonly used terms.
            Below 50: Poor – misaligned, incomplete, or non-standard.

    4) Justify the Rating: Provide a concise, evidence-based explanation in 1 sentence, citing specific strengths or shortcomings (e.g., missing critical skills, use of obscure terminology, or excellent alignment).

    5) Provide a Summary: Write a brief summary of the evaluation, highlighting key reasons for the rating and the points that is covered and not covered.

    6) Confidence Level: Rate your confidence in the evaluation from 1 to 100, where 100 indicates complete certainty in the rating and summary provided.

output structure:
{
    "rating" : integer // an integer value between 1 to 100,
    "summary": "sentence in a string" // concise, evidence-based explanation in 1 sentence
    "confidence" : integer // an integer value between 1 to 100
}
'''


def get_categories(job_description , job_skills):
    input_content = "job description:\n" + job_description + "\n\njob skills to be evaluated :\n" + job_skills
    response = client.beta.chat.completions.parse(
    model="gpt-4o-mini",
    temperature=0.5,
    # This is to enable JSON mode, making sure responses are valid json objects
    response_format=type_to_response_format_param(EvalStructure),
    messages=[
        {
            "role": "system",
            "content": evaluation_system_prompt
        },
        {
            "role": "user",
            "content": input_content
        }
    ],
    )

    return response.choices[0].message
```

### Creating the Batch File

The batch file should be in JSONL format and contain one line (JSON object) per request.

Request Format

Each request should be defined as follows:



{
  "custom_id": <REQUEST_ID>,
  "method": "POST",
  "url": "/v1/chat/completions",
  "body": {
    "model": <MODEL>,
    "messages": <MESSAGES>,
    // other parameters
  }
}

```python
# Creating an array of json tasks
from openai.lib._parsing._completions import type_to_response_format_param

tasks = []

for index, row in main_df[:10].iterrows():
    print(row)
    input_content = "job description:\n" + row["job_summary"] + "\njob skills:\n" + row["job_skills"]
    
    task = {
        "custom_id": f"task-{index}",
        "method": "POST",
        "url": "/v1/chat/completions",
        "body": {
            # This is what you would have in your Chat Completions API call
            "model": "gpt-4o-mini",
            "temperature": 0.5,
            "response_format": type_to_response_format_param(EvalStructure),
            "messages": [
                {
                    "role": "system",
                    "content": evaluation_system_prompt
                },
                {
                    "role": "user",
                    "content": input_content
                }
            ],
        }
    }
    
    tasks.append(task)
```

```python
# Creating the file

file_name = "batch_eval_simp_structured.jsonl"

def serialize_eval_structure(obj):
    if isinstance(obj, type(EvalStructure)):
        return str(obj)
    raise TypeError(f"Type {type(obj)} not serializable")

with open(file_name, 'w') as file:
    for obj in tasks:
        file.write(json.dumps(obj, default=serialize_eval_structure) + '\n')
```

```python
batch_file = client.files.create(
  file=open(file_name, "rb"),
  purpose="batch"
)
```

```python
def create_batch_tasks(df, model="gpt-4o-mini"):
    tasks = []
    
    for index, row in df.iterrows():
        input_content = f"job description:\n{row['job_summary']}\njob skills:\n{row['job_skills']}"
        
        task = {
            "custom_id": f"task-{index}",
            "method": "POST",
            "url": "/v1/chat/completions",
            "body": {
                "model": model,
                "temperature": 0.5,
                "response_format": type_to_response_format_param(EvalStructure),
                "messages": [
                    {"role": "system", "content": evaluation_system_prompt},
                    {"role": "user", "content": input_content}
                ],
            }
        }
        tasks.append(task)
    
    return tasks

# Create and save batch file
def save_batch_file(tasks, filename="batch_eval.jsonl"):
    with open(filename, 'w') as file:
        for task in tasks:
            file.write(json.dumps(task, default=lambda x: str(x)) + '\n')
    return filename
```

```python
def submit_batch_job(client, filename):
    # Upload file
    batch_file = client.files.create(
        file=open(filename, "rb"),
        purpose="batch"
    )
    
    # Create batch job
    batch_job = client.batches.create(
        input_file_id=batch_file.id,
        endpoint="/v1/chat/completions",
        completion_window="24h"
    )
    
    return batch_job

def monitor_batch_job(client, batch_job, check_interval=300):
    while True:
        batch_job = client.batches.retrieve(batch_job.id)
        if batch_job.status == "completed":
            print("Batch job completed!")
            break
        elif batch_job.status == "failed":
            print("Batch job failed!")
            break
        else:
            print(f"Batch job is still {batch_job.status}...")
            time.sleep(check_interval)
    
    return batch_job
```

```python
def process_batch_results(result_file_id, main_df):
    # Download results
    result = client.files.content(result_file_id).content
    
    # Process results into structured format
    results_list = []
    for res in json.loads(result):
        task_id = res['custom_id']
        index = int(task_id.split('-')[-1])
        result_content = json.loads(res['response']['body']['choices'][0]['message']['content'])
        
        result_dict = {
            'Task_ID': task_id,
            'Job_Summary': main_df.iloc[index]['job_summary'],
            'Job_Skills': main_df.iloc[index]['job_skills'],
            'rating': result_content.get('rating', 0),
            'summary': result_content.get('summary', ''),
            'confidence': result_content.get('confidence', 0)
        }
        results_list.append(result_dict)
    
    # Create DataFrame
    results_df = pd.DataFrame(results_list)
    results_df.set_index('Task_ID', inplace=True)
    
    return results_df
```

```python
def run_evaluation_pipeline(input_df, batch_size=100):
    # Split data into batches
    batches = np.array_split(input_df, len(input_df) // batch_size + 1)
    all_results = []
    
    for i, batch_df in enumerate(batches):
        print(f"Processing batch {i+1}/{len(batches)}")
        
        # Create and submit batch
        tasks = create_batch_tasks(batch_df)
        filename = save_batch_file(tasks, f"batch_{i}.jsonl")
        batch_job = submit_batch_job(client, filename)
        
        # Monitor and get results
        completed_job = monitor_batch_job(client, batch_job)
        batch_results = process_batch_results(completed_job.output_file_id, batch_df)
        
        all_results.append(batch_results)
    
    # Combine all results
    final_results = pd.concat(all_results)
    
    # Save results
    final_results.to_csv("evaluation_results.csv")
    
    return final_results
```

```python
run_evaluation_pipeline(main_df, batch_size=100)
```
